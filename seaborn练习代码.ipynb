```python
import seaborn as sns
```

```python
sns.set
```

```python
tips = sns.load_dataset('tips')
tips.head()
                        
```

```python
sns.relplot(x='total_bill', y="tip",data=tips, hue="smoker", size="size",style="smoker")
```

```python
dots = sns.load_dataset("dots")
```

```python
dots.head()
```

```python
sns.relplot(x="time", y="firing_rate", col="align",
           hue="choice", style="choice", size="coherence",
           facet_kws=dict(sharex=False),
           kind='line', legend="full", data=dots)
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(x="timepoint", y="signal", col="region",
            hue="event", style="event",
            kind="line", data=fmri);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="swarm", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="violin", split=True, data=tips);
sns.violin()
```

```python
import matplotlib.pyplot as plt
```

```python
f, axes = plt.subplots(1,2, sharey=True, figsize=(12,6))
sns.boxplot(x="day", y='tip', data=tips, ax=axes[0])
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, ax= axes[1])
```

```python
sns.relplot(x="time", y="firing_rate", col="align",
           hue="choice", size="coherence", style="choice",
           height=4.5, aspect=2/3,
           facet_kws=dict(sharex=False),
           kind="line", legend="full", data=dots);
```
